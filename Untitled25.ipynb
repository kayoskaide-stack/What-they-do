```python
# Single-cell: handle .res reference files, parse heuristically, compare with sample, and call OpenAI correctly.
import os, io, zipfile, json, csv, time, traceback, re
from collections import defaultdict
from IPython.display import display

def list_res_files_from_zip(zip_bytes):
    with zipfile.ZipFile(io.BytesIO(zip_bytes)) as z:
        return [n for n in z.namelist() if n.lower().endswith('.res')]

def parse_res_text(text):
    # Heuristic parser for .res-like text files: extract named numeric fields and common CELL parameters.
    lines = text.splitlines()
    data = {}
    # common crystallographic CELL line (a b c alpha beta gamma)
    for ln in lines:
        ln_strip = ln.strip()
        # key = value patterns
        if "=" in ln_strip:
            parts = [p.strip() for p in ln_strip.split("=" ,1)]
            if len(parts)==2:
                key,val = parts
                # keep numeric if possible
                nums = re.findall(r"[-+]?\d*\.\d+|\d+", val)
                if nums:
                    # store single number or list
                    if len(nums)==1:
                        data[key.lower()] = float(nums[0])
                    else:
                        data[key.lower()] = [float(x) for x in nums]
                else:
                    data[key.lower()] = val
                continue
        # CELL or cell line with multiple numeric entries
        m = re.match(r'^(CELL|cell)\s*[:=]?\s*(.*)', ln_strip)
        if m:
            vals = re.findall(r"[-+]?\d*\.\d+|\d+", m.group(2))
            if vals:
                data['cell'] = [float(x) for x in vals]
                continue
        # lines starting with a label and then numbers (e.g., "SCALE 1.000 0.000")
        m2 = re.match(r'^([A-Za-z0-9_+-]+)\s+((?:[-+]?\d*\.\d+\s+?)+)', ln_strip)
        if m2:
            key = m2.group(1).lower()
            nums = re.findall(r"[-+]?\d*\.\d+|\d+", m2.group(2))
            if nums:
                data[key] = [float(x) for x in nums] if len(nums)>1 else float(nums[0])
                continue
        # fallback: find any single numeric token labeled by preceding word
        m3 = re.match(r'^([A-Za-z_]+)\s+([-+]?\d*\.\d+|\d+)\b', ln_strip)
        if m3:
            data[m3.group(1).lower()] = float(m3.group(2))
    # also attempt to parse key:value patterns with colon
    for ln in lines:
        if ":" in ln:
            k,v = [p.strip() for p in ln.split(":",1)]
            if k.lower() not in data:
                nums = re.findall(r"[-+]?\d*\.\d+|\d+", v)
                if nums:
                    data[k.lower()] = [float(x) for x in nums] if len(nums)>1 else float(nums[0])
                else:
                    data[k.lower()] = v
    return data

# Access uploaded zip bytes from existing widget
if not zip_picker.value:
    raise RuntimeError("Please (re)upload Res_Database.zip in the zip_picker widget and re-run this cell.")

zip_bytes = list(zip_picker.value.values())[0]['content']
res_list = list_res_files_from_zip(zip_bytes)
print("Found .res files:", len(res_list))
for n in res_list[:50]:
    print(" -", n)

# Parse each .res into a reference entry
refs = []
with zipfile.ZipFile(io.BytesIO(zip_bytes)) as z:
    for name in res_list:
        try:
            text = z.read(name).decode('utf-8', errors='replace')
            parsed = parse_res_text(text)
            parsed['_source_file'] = name
            # collapse numeric lists by storing common metrics e.g., cell lengths mean, first number, etc.
            # Create a 'value' heuristic for matching: prefer cell a or first numeric any.
            candidate_value = None
            for k in ('cell','a','cell_length_a','length_a'):
                if k in parsed:
                    v = parsed[k]
                    if isinstance(v, list): candidate_value = v[0]
                    else: candidate_value = v
                    break
            # if no cell, pick first numeric field
            if candidate_value is None:
                for k,v in parsed.items():
                    if isinstance(v,(int,float)):
                        candidate_value = float(v); break
                    if isinstance(v,list) and all(isinstance(x,(int,float)) for x in v):
                        candidate_value = float(v[0]); break
            parsed['value'] = candidate_value
            refs.append(parsed)
        except Exception as e:
            print("Failed to parse", name, e)

print("Parsed", len(refs), "reference entries with heuristic 'value' fields.")
# If no reference numeric values found, report keys commonly present
if not refs:
    raise RuntimeError("No .res references parsed — confirm .res files are text and not compressed differently.")
if all(r.get('value') is None for r in refs):
    print("Warning: parser found no clear numeric 'value' fields. Sample of parsed keys from first file:")
    print(list(refs[0].keys()))

# Now load sample measurements from sample_picker widget (like earlier)
if not sample_picker.value:
    raise RuntimeError("Please (re)upload results_summary.json in the sample_picker widget and re-run this cell.")

sample_bytes = list(sample_picker.value.values())[0]['content']
sample_json = json.loads(sample_bytes.decode('utf-8'))
# find measurements list heuristically
if isinstance(sample_json, dict) and "measurements" in sample_json:
    measurements = sample_json["measurements"]
elif isinstance(sample_json, list):
    measurements = sample_json
else:
    def find_list(obj):
        if isinstance(obj, list): return obj
        if isinstance(obj, dict):
            for v in obj.values():
                r = find_list(v)
                if r: return r
        return None
    measurements = find_list(sample_json) or []
if not measurements:
    raise RuntimeError("Could not find measurement list in results_summary.json; inspect its structure.")

def normalize_entry(e):
    out = {}
    if isinstance(e, dict):
        for k,v in e.items():
            lk = k.lower()
            if lk in ("name","id","label","filename","sample"):
                out['name'] = str(v)
            if lk in ("value","measurement","meas","val","result"):
                try:
                    out['value'] = float(v)
                except:
                    out['value'] = None
            if lk in ("units","unit"):
                out['units'] = str(v)
            out.setdefault('raw',{})[k]=v
    else:
        out['raw']=e
    out['name'] = out.get('name') or out['raw'].get('id') if isinstance(out.get('raw'),dict) else 'sample'
    return out

samples = [normalize_entry(e) for e in measurements]
print("Loaded", len(samples), "sample entries. Sample values:", [s['value'] for s in samples[:10]])

# Matching function: match sample.value to refs' 'value' by nearest numeric distance
def match_by_numeric(sample_val, refs, top_n=5):
    if sample_val is None:
        return []
    cand = []
    for r in refs:
        rv = r.get('value')
        if rv is None: continue
        try:
            diff = abs(float(sample_val) - float(rv))
            cand.append((diff, r))
        except:
            continue
    cand.sort(key=lambda x:x[0])
    return [r for d,r in cand[:top_n]]

report_rows = []
for s in samples:
    sval = s.get('value')
    row = {'sample_name': s.get('name','sample'), 'sample_value': sval, 'ref_matches': []}
    if sval is not None:
        matches = match_by_numeric(sval, refs, top_n=5)
        for m in matches:
            row['ref_matches'].append({'_source_file': m.get('_source_file'), 'value': m.get('value'), 'all_keys': {k:v for k,v in m.items() if k!='_source_file'}})
    report_rows.append(row)

# Write CSV summary
out_csv = "results_comparison_res.csv"
with open(out_csv, "w", newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['sample_name','sample_value','top_ref_file','ref_value'])
    for r in report_rows:
        if r['ref_matches']:
            for m in r['ref_matches']:
                writer.writerow([r['sample_name'], r['sample_value'], m['_source_file'], m['value']])
        else:
            writer.writerow([r['sample_name'], r['sample_value'], '', ''])
print("Wrote", out_csv)
print("Sample report rows:")
for r in report_rows[:10]:
    print(r)

# Optional: call OpenAI for a high-level summary (use proper call signature)
if key_widget.value and key_widget.value.strip():
    try:
        from openai import OpenAI
        client = OpenAI(api_key=key_widget.value.strip())
        prompt_lines = ["You are an expert materials scientist. Compare these sample measurements to reference .res entries and advise likely crystal types, confidence, and next steps. Note instrument/optics limitations."]
        for r in report_rows[:40]:
            prompt_lines.append(str(r))
        prompt = "\n".join(prompt_lines)
        # call with backoff, using Responses.create without unsupported params
        def call_openai(prompt_text, max_retries=5):
            backoff=1.0
            for i in range(max_retries):
                try:
                    resp = client.responses.create(model="gpt-4o-mini", input=prompt_text)
                    # extract text from resp.output
                    out_text = ""
                    if hasattr(resp, "output") and resp.output:
                        for item in resp.output:
                            if isinstance(item, dict) and "content" in item:
                                c = item["content"]
                                if isinstance(c, list):
                                    for piece in c:
                                        if piece.get("type")=="output_text":
                                            out_text += piece.get("text","")
                                elif isinstance(c, str):
                                    out_text += c
                    if not out_text:
                        out_text = str(resp)
                    return out_text
                except Exception as e:
                    msg=str(e)
                    print(f"OpenAI call failed attempt {i+1}:", msg)
                    if "quota" in msg.lower() or "insufficient_quota" in msg.lower():
                        raise
                    time.sleep(backoff); backoff*=2
            raise RuntimeError("OpenAI call failed after retries")
        print("\nCalling OpenAI for analysis...")
        analysis = call_openai(prompt)
        print("\n--- OpenAI analysis ---\n")
        print(analysis)
    except Exception as e:
        print("OpenAI call failed:", e)
        traceback.print_exc()
else:
    print("No API key provided or empty; skipping OpenAI analysis.")
```
